Preparando o ambiente de desenvolvimento

```python
# Observar o ambiente do interpretador python. no meu caso, tenho o python instalado em dois ambientes diferentes, 
# em um deles não consegui importar o modulo statsmodels, e não consegui instalalo por ja esta instalado 
# no outro ambiente. Portanto tive de selecionar o interpretador no ambiente espefico.


import pandas as pd

from statsmodels.tsa.seasonal import seasonal_decompose

import matplotlib.pyplot as plt

import seaborn as sns
```

--------

Tratando o arquivo 

```python
caminho = 'pim.xlsx' #caminho do arquivo
```

```python
df = pd.read_excel(caminho) #carregando o arquivo com base no caminho.
```

```python
#Visualizando informações do arquivo.

print(df.head())#Primeiras linhas do Data Frame
print(df.tail())#Ultimas linhas do Data Frame
print(df.info())#Informações uteis.
```

----------
Analise da serie temporal.

```python
plt.plot(df.Month, df.Value)#Plota grafico da serie Produção Industrial mensal  PIM - PF
```

Decomposição de acordo com o modelo aditivo

```python
#Antes é necessario transformar a data da serie em indice. 
# Serve para os dois modelos, aditivo e multiplicativo

df.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

```python
#Decomposição

resultado = seasonal_decompose(df, model='additive')
```

```python
#Plotando o resultado

resultado.plot()
plt.show
```

Decompondo de acordo com o modelo multiplicativo

```python
resultado2 = seasonal_decompose(df, model='multiplicative')
```

```python
resultado2.plot()
plt.show()
```
